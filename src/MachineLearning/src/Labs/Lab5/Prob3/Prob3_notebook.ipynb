```python
get_ipython().magic('matplotlib notebook')

import matplotlib.pyplot as pl
import numpy as np
```

```python
p = np.array([[2,2],[2,-2],[-2,-2],[-2,2]])
n = np.array([[1,1],[1,-1],[-1,-1],[-1,1]])

pl.plot(p[:,0],p[:,1],'ro')
pl.plot(n[:,0],n[:,1],'bx')
pl.show()
```

```python
def phi(x):
    if np.transpose(x).dot(x) > 4:
        k = abs(x[0] - x[1])
        return np.array([4-x[1]+k,4-x[0]+k])
    else:
        return x
```

```python
def phis(x):
    return np.reshape([phi(x[i]) for i in range(np.shape(x)[0])], (4,2))
```

```python
pp = phis(p)
nn = phis(n)

print(pp)
print(nn)
```

```python
pl.figure()
pl.plot(pp[:,0],pp[:,1],'ro')
pl.plot(nn[:,0],nn[:,1],'bx')
pl.plot([-1, 4], [4, -1], 'k-')
pl.show()
```
